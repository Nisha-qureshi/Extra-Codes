```python
def speak(audio):
    engine = pyttsx3.init()
    engine.say(audio)
    engine.runAndWait()
#speak("hii")    
    
#import datetime    
def time():
    Time = datetime.datetime.now().strftime("%I:%M:%S")
    speak(Time)
    print(Time)
# time()

def date():
    year = int(datetime.datetime.now().year)
    month = int(datetime.datetime.now().month)
    date = int(datetime.datetime.now().day)
    speak(date)
    print(date)
    speak(month)
    print(month)
    speak(year)
    print(year)
#date()

def wishme():
    speak("Welcome back Nisha")
    print("Welcome back Nisha")
    speak("The current time is")
    print("The current time is")
    time()
    speak("The current date is ")
    print("The current date is ")
    date()
    speak("Jarvis at your service Please tell me how can i help you")
    print("Jarvis at your service Please tell me how can i help you")
    hour = datetime.datetime.now().hour
    
    if hour >=6 and hour<12:
        speak("Good Morning Nisha!")
        print("Good Morning Nisha!")

    elif hour >=12 and hour<18:
        speak("Good Afternoon Nisha!")
        print("Good Afternoon Nisha!")
        
    elif hour >=18 and hour<24:
        speak("Good Evening Nisha!")
        print("Good Evening Nisha!")
        
    else:
        speak("Good Night Nisha!")
        print("Good Night Nisha!")
# wishme()

# import psutil
def cpu():
    usage = str(psutil.cpu_percent())
    speak("CPU Usage"+usage)
    print("CPU Usage"+usage)
    battery = psutil.sensors_battery()
    speak("Battery percent")
    speak(battery.percent)
    print("Battery")
    print(battery.percent)

# cpu()

#import pyautogui
def screenshot():
    img=pyautogui.screenshot()
    img.save("G://nisha/jarvis.png")

# screenshot()

# play songs
def playSong():
    mdir="G:\\nisha\\nis_songs"
    songs=os.listdir(mdir)
    os.startfile(os.path.join(mdir, songs[4]))
#playSong()

def jokes():
    speak(pyjokes.get_joke())
    
jokes()
```

```python
#import pyaudio

def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening.....")
        r.pause_threshold = 1 #waiting for 1s and then start listening
        audio = r.listen(source)
        
    try:
        print("Recognizing...")
        query = r.recognize_google(audio, language="en-in")
        print(query)

    except Exception as e:
        print(e)
        speak("Say that again please...")
        print("Say that again please...")
        
            
        return "None"
    return query
    
takeCommand()
```

```python
if __name__=="__main__":
    wishme()
    while True:
        query = takeCommand().lower()
        
        if 'time' in query:
            time()
            
        elif 'date' in query:
            date()
            
        elif 'wikipedia' in query:
            speak('Searching ....')
            query = query.replace("wikipedia","")
            result = wikipedia.summary(query, sentences=2)
            print(result)
            speak(result)
            
        elif 'remember that' in query:
            speak("Hello nisha what should i remember")
            data=takeCommand()
            speak("you have said "+data+"to remember")
            remember=open("jarvis.txt","w")
            remember.write(data)
            remember.close()
            
        elif 'Do you know anything' in query:
            remember=open("jarvis.txt","r")
            speak("you have said me to remember"+remember.read())
            
        elif 'screenshot' in query:
            screenshot()
            speak("Screenshot has been successfully saved!")
            print("Screenshot has been successfully saved!")
        
        elif 'cpu' in query:
            cpu()
            
        elif 'logout' in query:
            os.system("shutdown -l")
       
        elif 'shutdown' in query:
            os.system("shutdown /s /t -l")
        
        elif 'restart' in query:
            os.system("shutdown /r /t -l")
       
        elif 'joke' in query:
            jokes()
        
        elif  'play songs' in query:
            playSong()
            
        elif 'offline' in query:
            quit()
            
        
```
