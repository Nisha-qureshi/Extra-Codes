```python
from PIL import Image, ImageDraw, ImageFont
image = Image.new('RGB', (1000,900), (255, 255, 255))
draw = ImageDraw.Draw(image)
font = ImageFont.truetype('arial.ttf', size=45)
import random
import os
import datetime
import qrcode
os.system("ID CARD Generator by NISHA and SAPNA")

from tkinter import *
import os

credit="id_card.temp"

def id_card():           
    d_date = datetime.datetime.now()
    reg_format_date = d_date.strftime("  %d-%m-%Y\t\t\t\t\t ID CARD Generator\t\t\t\t\t  %I:%M:%S %p")
    print ('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    print (reg_format_date)
    print ('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')

    # starting position of the message
    print('\n\nAll Fields are Mandatory') 
    print('Avoid any kind of Spelling Mistakes')
    print('Write Everything in uppercase letters')
    (x, y) = (50, 50)
    def start():
        type_msg=int(input("\nChoose an option for whom you want to create id card:\n1.Employee.....\n2.Faculty....\n3.Student....\n\n"))
        if type_msg==1:
            message = input('\nEnter Your Company Name: ')
        elif type_msg==2 or type_msg==3:
            message = input('\nEnter Your College Name: ')
        else:
            print("\nPlease choose an correct option")
            start()

    company=start()
    color = 'rgb(0, 0, 0)'
    font = ImageFont.truetype('arial.ttf', size=80)
    draw.text((x, y), company, fill=color, font=font)





    # adding an unique id number. You can manually take it from user
    (x, y) = (600, 75)
    idno=random.randint(10000000,90000000)
    message = str('ID '+str(idno))
    color = 'rgb(0, 0, 0)' # black color
    font = ImageFont.truetype('arial.ttf', size=60)
    draw.text((x, y), message, fill=color, font=font)




    (x, y) = (50, 250)
    message = input('Enter Your Full Name: ')
    name=message
    color = 'rgb(0, 0, 0)' # black color
    font = ImageFont.truetype('arial.ttf', size=45)
    draw.text((x, y), message, fill=color, font=font)



    (x, y) = (50, 350)
    message = input('Enter Your Gender: ')
    color = 'rgb(0, 0, 0)' # black color 
    draw.text((x, y), message, fill=color, font=font)
    (x, y) = (250, 350)
    message = input('Enter Your Age: ')
    color = 'rgb(0, 0, 0)' # black color 
    draw.text((x, y), message, fill=color, font=font)



    (x, y) = (50, 450)
    message = input('Enter Your Date Of Birth: ')
    color = 'rgb(0, 0, 0)' # black color 
    draw.text((x, y), message, fill=color, font=font)



    (x, y) = (50, 550)
    message = input('Enter Your Blood Group: ')
    color = 'rgb(255, 0, 0)' # black color 
    draw.text((x, y), message, fill=color, font=font)



    (x, y) = (50, 650)
    message = input('Enter Your Mobile Number: ')
    temp=message
    color = 'rgb(0, 0, 0)' # black color 
    draw.text((x, y), message, fill=color, font=font)




    (x, y) = (50, 750)
    message = input('Enter Your Address: ')
    color = 'rgb(0, 0, 0)' # black color 
    draw.text((x, y), message, fill=color, font=font)




    # save the edited image

    image.save(str(name)+'.png')



    img = qrcode.make(str(company)+str(idno))   # this info. is added in QR code, also add other things
    img.save(str(idno)+'.bmp')


    til = Image.open(name+'.png')
    im = Image.open(str(idno)+'.bmp') #25x25
    til.paste(im,(600,350))
    til.save(name+'.png')

    print(('\n\n\nYour ID Card Successfully created in a PNG file '+name+'.png'))

def Signup():
    global window
    global nameE
    global passwordE
    window=Tk()
    window.title("ID card generator")
    window.geometry("500x200")

    l1=Label(window,text='Please enter new credentials:\n')
    l1.grid(row=0,column=0,sticky=E)

    nameL=Label(window,text="Username")
    nameL.grid(row=1,column=0,sticky=W)

    passwordL=Label(window,text="Password")
    passwordL.grid(row=2,column=0,sticky=W)

    nameE=Entry(window)
    passwordE=Entry(window,show="*")
    
    nameE.grid(row=1,column=1)
    passwordE.grid(row=2,column=1)
    
    sign_upb=Button(window,text="sign up",command=FSSignup)
    sign_upb.grid(columnspan=2,sticky=W)
    window.mainloop()

def FSSignup():
    with open(credit,"w") as f:
        f.write(nameE.get())
        f.write("\n")
        f.write(passwordE.get())
        f.close()
    window.destroy()
    Login()

def Login():
    global nameEL
    global passwordEL
    global login_window
    
    login_window=Tk()
    login_window.title("Login")
    login_window.geometry("300x200")
    l2=Label(login_window,text="Please log in")
    l2.grid(sticky=E)
    
    nameL=Label(login_window,text="Username : ")
    nameL.grid(row=1,column=0,sticky=W)

    passwordL=Label(login_window,text="Password : ")
    passwordL.grid(row=2,column=0,sticky=W)

    nameEL=Entry(login_window)
    passwordEL=Entry(login_window,show="*")
    
    nameEL.grid(column=1,row=1)
    passwordEL.grid(column=1,row=2)
    
    log_inb=Button(login_window,text="Login",command=Checklogin)
    log_inb.grid(columnspan=2,sticky=W)
    
    rmuser=Button(login_window,text="Delete user",command=DelUser)
    rmuser.grid(columnspan=2,sticky=W)
    login_window.mainloop()

def Checklogin():
    with open(credit) as f:
        data=f.readlines()
        uname=data[0].rstrip()
        password=data[1].rstrip()
        
        if nameEL.get()==uname and passwordEL.get()==password:
            r=Tk()
            r.title("D:")
            r.geometry("200x150")
            rlb1=Label(r,text="\n [+] Logged In")
            rlb1.pack()
            r.mainloop()
        else:
            r=Tk()
            r.title("D:")
            r.geometry("200x150")
            rlb1=Label(r,text="\n [!] Invalid Login")
            rlb1.pack()
            r.mainloop()
            
def DelUser():
    os.remove(credit) #credit="id_card.tmp"
    login_window.destroy()  #login_window=tkinter.Tk()
    Sign_up()

if os.path.isfile(credit):
    Login()
    id_card()
else :
    Signup()
    id_card()

```
